# 3.1.2 Hypothesis Testing: Comparing Two Graphs

```python
from scipy import stats #let's do some statistics!
```

## 3.1.2.1 Student's t-test: the simplest statistical test

### 1-sample t-test: testing the value of a population mean

```python
stats.ttest_1samp(data["VIQ"], 0) 
#tests if the population mean of data is likely to be equal to a given value
#I couldn't figure out how to get rid of np.float64 in this case... sad
```

### 2-sample t-test: testing for difference across populations

```python
female_VIQ = data[data["Gender"] == "Female"]["VIQ"]
male_VIQ = data[data["Gender"] == "Male"]["VIQ"]
stats.ttest_ind(female_VIQ, male_VIQ)
#We can test if the difference in mean VIQ between male and female populations is significant!
#np.float64.........sad again
```

## 3.1.2.2 Paired tests: repeated measurements on the same individuals

```python
stats.ttest_ind(data["FSIQ"], data["PIQ"])
#Are FSIQ and PIQ significantly different? Let's see via 2 sample test
```

```python
stats.ttest_rel(data["FSIQ"], data["PIQ"])
#this tests a paired relationship since FSIQ and PIQ values are measured on the same individuals
```

```python
stats.ttest_1samp(data["FSIQ"] - data["PIQ"], 0)
#the above example is equivalent to 1-sample test on the difference. Fun stuff
```

```python
stats.wilcoxon(data["FSIQ"], data["PIQ"])
#t-tests assume Gaussian distrubution, the Wilcoxon test does not
```

### Exercise 3 - 3.1.2

Test the difference between weights and males and females

```python
female_weight = data[data["Gender"] == "Female"]["Weight"].dropna() #drop N/A values
male_weight = data[data["Gender"] == "Male"]["Weight"].dropna() #drop N/A values
stats.ttest_ind(female_weight, male_weight) #run 2 sample t-test of female vs male weights
```

Use non parametric statistics to test the difference between VIQ in males and females

```python
f_VIQ = data[data["Gender"] == "Female"]["VIQ"] #define female VIQ
m_VIQ = data[data["Gender"] == "Male"]["VIQ"] #define male VIQ
stats.mannwhitneyu(f_VIQ, m_VIQ) #Mann Whitney test of female vs male VIQ, nonparametric
```

The data does not support the hypothesis that there is a difference in VIQ between females and males

# 3.1.3 Linear models, multiple factors, and analysis of variance

## 3.1.3.1 “Formulas” to specify statistical models in Python

### A simple linear regression

```python
x = np.linspace(-5, 5, 20) #create NumPy array
np.random.seed(1) #set seed for reproducibility
y = -5 + 3*x + 4 * np.random.normal(size=x.shape) #generate noisy linear data
data = pd.DataFrame({'x': x, 'y': y}) #create data frame from x and y
```

```python
from statsmodels.formula.api import ols #import ols for linear models
model = ols("y ~ x", data).fit() #fit a linear regression model of y on x
```

```python
print(model.summary()) #print the summary of the model
```

### Exercise 4 - 3.1.3

Retrieve the estimated parameters from the model above. Hint: use tab-completion to find the relevent attribute

```python
intercept = model.params['Intercept'] #define y-int
slope = model.params['x'] #define slope
print("y-int is:", intercept) #print intercept
print("slope is:", slope) #print slope
```

### Categorical variables: comparing groups or multiple categories

```python
data = pd.read_csv('brain_size.csv', sep=';', na_values=".") #redefine dataset to brain_size.csv
```

```python
model = ols("VIQ ~ Gender + 1", data).fit() #fit linear model predicting VIQ by Gender with intercept
print(model.summary())  
```

```python
model = ols('VIQ ~ C(Gender)', data).fit() #forcing integer column to be treated as categorical
```

```python
data_fisq = pd.DataFrame({'iq': data['FSIQ'], 'type': 'fsiq'}) #make long data
data_piq = pd.DataFrame({'iq': data['PIQ'], 'type': 'piq'})
data_long = pd.concat((data_fisq, data_piq)) #define long data
print(data_long) #print long data
```

```python
model = ols("iq ~ type", data_long).fit() #fit linear regression of iq predicted by type
print(model.summary())  
```

```python
stats.ttest_ind(data['FSIQ'], data['PIQ']) #2-sample t-test again
```

## 3.1.3.2 Multiple Regression: including multiple factors

```python
data = pd.read_csv('iris.csv') #now we are using iris.csv
model = ols('sepal_width ~ name + petal_length', data).fit() #fit linear model predicting sepal_width from name and petal_length
print(model.summary())  
```

## 3.1.3.3 Post-hoc hypothesis testing: analysis of variance (ANOVA)

```python
print(model.f_test([0, 1, -1, 0])) #test if coefficients for second and third predictors are equal
```

### Exercise 5 - 3.1.3

Going back to the brain size + IQ data, test if the VIQ of male and female are different after removing the effect of brain size, height and weight

```python
# Import statsmodels again
import statsmodels.formula.api as smf
# Redefine dataset
data = pd.read_csv("brain_size.csv", sep=";", na_values=".")
# Separate independent variables
VIQ_model = smf.ols('VIQ ~ Gender + MRI_Count + Height + Weight', data=data).fit()
print(VIQ_model.summary())
```

# 3.1.4 More visualization: seaborn for statistical exploration

```python
import urllib.request
import os
```

```python
if not os.path.exists('wages.txt'):
    # Download the file if it is not present
    urllib.request.urlretrieve('http://lib.stat.cmu.edu/datasets/CPS_85_Wages',
                       'wages.txt')

# Give names to the columns
names = [
    'EDUCATION: Number of years of education',
    'SOUTH: 1=Person lives in South, 0=Person lives elsewhere',
    'SEX: 1=Female, 0=Male',
    'EXPERIENCE: Number of years of work experience',
    'UNION: 1=Union member, 0=Not union member',
    'WAGE: Wage (dollars per hour)',
    'AGE: years',
    'RACE: 1=Other, 2=Hispanic, 3=White',
    'OCCUPATION: 1=Management, 2=Sales, 3=Clerical, 4=Service, 5=Professional, 6=Other',
    'SECTOR: 0=Other, 1=Manufacturing, 2=Construction',
    'MARR: 0=Unmarried,  1=Married',
]

short_names = [n.split(':')[0] for n in names]

data = pd.read_csv('wages.txt', skiprows=27, skipfooter=6, sep=None,
                       header=None, engine = "python")
data.columns = short_names

# Log-transform the wages, because they typically are increased with
# multiplicative factors
import numpy as np
data['WAGE'] = np.log10(data['WAGE'])
```

```python
import seaborn
```

```python
print(data)
```

### 3.1.4.1. Pairplot: scatter matrices

```python
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'],
                 kind='reg') 
#create pairplot scatter matrix for wage, age, and education
```

```python
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'],
                 kind='reg', hue='SEX') 
#same thing, but we are separating gender data by different colors
```

### 3.1.4.2. lmplot: plotting a univariate regression

```python
seaborn.lmplot(y='WAGE', x='EDUCATION', data=data) 
#linear model showing relationip between wage and education
```
